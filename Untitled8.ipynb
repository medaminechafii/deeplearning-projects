```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples=10000,noise=0.1,factor=0.3,random_state=0)
print(X.shape)
print(y.shape)
y = y.reshape(-1,1)
```

```python
plt.scatter(X[:,0],X[:,1], c = y)
```

```python
X=X.T
y=y.T
print(X.shape)
print(y.shape)
```

```python
def init (dimensions):
  depth = len(dimensions)
  parameters = {}
  for i in range(1,depth):
    parameters["W"+str(i)] = np.random.randn(dimensions[i],dimensions[i-1])
    parameters["b"+str(i)] = np.zeros((dimensions[i],1))
  return parameters
```

```python
def forward_propagation (X,parameters):
  depth = len(parameters)//2
  activations ={"A0":X}
  for i in range(1,depth+1):
    W = parameters["W"+str(i)]
    b = parameters["b"+str(i)]
    A = activations["A"+str(i-1)]
    Z = np.dot(W,A) + b
    A = 1/(1+np.exp(-Z))
    activations["A"+str(i)] = A
  return activations
```

```python
def logloss(A,y):
  return -np.mean(y*np.log(A) + (1-y)*np.log(1-A))
```

```python
def back_propagation(X,y,activations,parameters):
  m = y.shape[1]
  C = len(parameters)//2

  gradients ={}
  dZ = activations["A"+str(C)] - y
  for i in range(C,0,-1):
    W = parameters["W"+str(i)]
    A = activations["A"+str(i-1)]
    dW = np.dot(dZ,A.T)/m
    db = np.sum(dZ,axis=1,keepdims=True)/m
    if i >1:
      dZ = np.dot(W.T,dZ)*A*(1-A)
    gradients["dW"+str(i)] = dW
    gradients["db"+str(i)] = db
  return gradients
```

```python
def update(gradients,parameters,alpha):
  C = len(parameters)//2
  for i in range(1,C+1):
    parameters["W"+str(i)] = parameters["W"+str(i)] - alpha*gradients["dW"+str(i)]
    parameters["b"+str(i)] = parameters["b"+str(i)] - alpha*gradients["db"+str(i)]
  return parameters
```

```python
def predict (X,parameters):
  C = len(parameters)//2
  A = forward_propagation(X,parameters)["A"+str(C)]
  return A>=0.5
```

```python
from tqdm import tqdm
```

```python
def neural_network(X_train,y_train,hidden_layers=[32,32,32],alpha=0.1,epoch = 1000):
  dimensions = list(hidden_layers)
  dimensions.insert(0,X_train.shape[0])
  dimensions.append(y_train.shape[0])
  parameters = init(dimensions)
  C = len(parameters)//2
  train_loss = []
  train_acc = []
  for i in tqdm(range(epoch)):
    activations = forward_propagation(X_train,parameters)
    gradients = back_propagation(X_train,y_train,activations,parameters)
    parameters = update(gradients,parameters,alpha)
    if i%10 == 0:
      y_pred = predict(X_train,parameters)
      loss = logloss(activations["A"+str(C)],y_train)
      train_loss.append(loss)
      current_accuracy = np.mean(y_pred == y_train)
      train_acc.append(current_accuracy)
  print("Accuracy: ",np.mean(y_pred == y))
  plt.figure(figsize=(10,5))
  plt.subplot(1,2,1)
  plt.plot(train_loss)
  plt.legend()
  plt.subplot(1,2,2)
  plt.plot(train_acc)
  plt.legend()
  plt.show()
  return parameters
```

```python
parameters = neural_network(X,y,hidden_layers=[32,32,32],epoch = 1000,alpha = 0.1)
```

```python
y_pred = predict(X,parameters)
print(np.mean(y_pred==y))
```
