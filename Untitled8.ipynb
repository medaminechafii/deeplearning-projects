```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples=1000,noise=0.1,factor=0.3,random_state=0)
print(X.shape)
print(y.shape)
y = y.reshape(-1,1)
```

```python
plt.scatter(X[:,0],X[:,1], c = y)
```

```python
X=X.T
y=y.T
print(X.shape)
print(y.shape)
```

```python
def init (n0,n1,n2):
  W1 = np.random.randn(n1,n0)
  b1 = np.random.randn(n1,1)
  W2 = np.random.randn(n2,n1)
  b2 = np.random.randn(n2,1)
  parameters = {"W1":W1,"b1":b1,"W2":W2,"b2":b2}
  return parameters
```

```python
def forward_propagation (X,parameters):
  W1 = parameters["W1"]
  b1 = parameters["b1"]
  W2 = parameters["W2"]
  b2 = parameters["b2"]
  Z1 = np.dot(W1,X) + b1
  A1= 1/(1+np.exp(-Z1))
  Z2 = np.dot(W2,A1) + b2
  A2 = 1/(1+np.exp(-Z2))
  activations = {"A1":A1,"A2":A2}
  return activations
```

```python
def logloss(A,y):
  return -np.mean(y*np.log(A) + (1-y)*np.log(1-A))
```

```python
def back_propagation(X,y,activations,parameters):
  m = y.shape[1]
  A1 = activations["A1"]
  A2 = activations["A2"]
  W2 = parameters["W2"]
  dZ2 = A2 - y
  dW2 = np.dot(dZ2,A1.T)/m
  db2 = np.sum(dZ2,axis=1,keepdims=True)/m
  dZ1 = np.dot(W2.T,dZ2)*A1*(1-A1)
  dW1 = np.dot(dZ1,X.T)/m
  db1 = np.sum(dZ1,axis=1,keepdims=True)/m
  gradients = {"dW1":dW1,"db1":db1,"dW2":dW2,"db2":db2}
  return gradients
```

```python
def update(gradients,parameters,alpha):
  W1 = parameters["W1"]
  b1 = parameters["b1"]
  W2 = parameters["W2"]
  b2 = parameters["b2"]
  dW1 = gradients["dW1"]
  db1 = gradients["db1"]
  dW2 = gradients["dW2"]
  db2 = gradients["db2"]
  W1 = W1 - alpha*dW1
  b1 = b1 - alpha*db1
  W2 = W2 - alpha*dW2
  b2 = b2 - alpha*db2
  parameters = {"W1":W1,"b1":b1,"W2":W2,"b2":b2}
  return parameters
```

```python
def predict (X,parameters):
  A = forward_propagation(X,parameters)["A2"]
  return A>=0.5
```

```python
def neural_network(X_train,y_train,n1,alpha=0.1,epoch = 1000):
  n0 = X_train.shape[0]
  n2 = y_train.shape[0]
  parameters = init(n0,n1,n2)
  train_loss = []
  train_acc = []
  for i in range(epoch):
    activations = forward_propagation(X_train,parameters)
    gradients = back_propagation(X_train,y_train,activations,parameters)
    parameters = update(gradients,parameters,alpha)
    if i%10 == 0:
      y_pred = predict(X_train,parameters)
      loss = logloss(activations["A2"],y_train)
      train_loss.append(loss)
      current_accuracy = np.mean(y_pred == y_train)
      train_acc.append(current_accuracy)
  print("Accuracy: ",np.mean(y_pred == y))
  plt.figure(figsize=(10,5))
  plt.subplot(1,2,1)
  plt.plot(train_loss)
  plt.legend()
  plt.subplot(1,2,2)
  plt.plot(train_acc)
  plt.legend()
  plt.show()
  return parameters
```

```python
parameters = neural_network(X,y,n1=8,epoch = 10000,alpha = 0.01)
```

```python
y_pred = predict(X,parameters)
print(np.mean(y_pred==y))
```
